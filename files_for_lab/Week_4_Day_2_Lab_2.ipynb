```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_Cat = pd.read_csv("categorical.csv")
data_Cat.columns
```

```python
data_Cat.drop(['Unnamed: 0', 'customer'],axis =1,inplace =True)
```

```python
data_Cat.columns
```

```python
# Plot all the categorical variables with the proper plot. What can you see?
def plot(x):
    for i in x.columns:
        sns.countplot(data_Cat[i])
        plt.xlabel(i)
        plt.show()
plot(data_Cat)
```

```python
#There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?
for i in data_Cat.columns:
    display(i)
    display(data_Cat[i].value_counts())
```

```python
# we can merge Luxury SUV and Luxury Car in one category from vehicle_class
def luxury_Car(x):
    if x in ("Luxury SUV","Luxury Car"):
        return "Luxury"
    else:
        return x
data_Cat["vehicle_class"] = list(map(luxury_Car,data_Cat["vehicle_class"])) 
 
```
